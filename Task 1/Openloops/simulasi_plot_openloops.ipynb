```python
import pandas as pd
import matplotlib.pyplot as plt

# Membaca file CSV
df_IL = pd.read_csv("IL_openloops.csv")
df_Vo = pd.read_csv("Vo_openloops.csv")

# Ubah nama kolom agar lebih mudah digunakan
df_IL.columns = ['Time', 'Current_IL']
df_Vo.columns = ['Time', 'Voltage_Vo']

# Plot arus IL
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(df_IL['Time'], df_IL['Current_IL'], label='IL (A)', color='blue')
plt.xlabel('Time (s)')
plt.ylabel('Current (A)')
plt.title('Inductor Current vs Time')
plt.grid(True)
plt.legend()

# Plot tegangan Vo
plt.subplot(1, 2, 2)
plt.plot(df_Vo['Time'], df_Vo['Voltage_Vo'], label='Vo (V)', color='green')
plt.xlabel('Time (s)')
plt.ylabel('Voltage (V)')
plt.title('Output Voltage vs Time')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```
